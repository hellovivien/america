**Question 1**

*Pour les travailleurs de la data aux Etats Unis donnez:*
* le salaire moyen, le salaire median, et représentez les dix déciles. Faites l'exercice pour les deux bases et comparez les résultats.
* Le salaire moyen en fonction du job title + diagramme circulaire de la répartitions de ses jobtitles (base 1)

-------------------------------

```python
def q1(df_name, df):
    line(df_name)
    print("salaire moyen : ", round(df["salary"].mean()))
    print("salaire median : ", df["salary"].median())
    print("quantiles : ", [df["salary"].quantile(i*0.1) for i in range(0,10)])

    salary_mean_per_job = df.groupby(["job_title"]).salary.mean().sort_values()
    md("**Salaire moyen par job**")
    display(salary_mean_per_job)
    
    job_values = df["job_title"].value_counts()
    plt.figure(figsize=(10, 10))
    plt.title("Diagramme circulaire de la répartition des jobs")
    job_values.plot.pie(labeldistance = 1.2, pctdistance = 0.8, autopct="%.1f%%", shadow = True, explode = [0.05 for _ in range(job_values.size)])
```

```python
q1(analyst_name, analyst_df_clean)
```

```python
q1(survey_name, survey_df_clean)
```

**Comparaison**

**La base *Analyst* a un salaire moyen plus faible, cela signifie que certains métiers représentés dans la base *Survey* sont mieux payés notemment *Architect*, *Data Scientist* et *Manager*. Le métier d'Analyste est le moins rémunérateur ce qui est confirmé par les premiers quantiles très proches dans les deux bases. On remarque que le salaire des Data Analystes seniors est inférieure à la moyenne dans la base *Analyst* cela suggère que pour un plan de carrière il faut mieux évoluer vers d'autres métiers de la Data que de devenir senior Analyste. Data Scientist semble être la meilleure porte d'entrée.**

**Question 9 + Heat-map**

*Que pensez vous de mon hypothèse HO: "Les personnes bien payées ont tendance à vouloir rester dans leur entreprise.*

```python
df["lf_jobs"].value_counts()
display(df.groupby(["lf_jobs"]).salary.mean())
df.loc[(df["lf_jobs"] == "Not Asked"), "lf_jobs"] = None
g = sns.catplot(x=df.lf_jobs, y=df.salary, data=df, aspect=2);
g.fig.suptitle("Répartition des salaires par rapport aux intentions de quitter son job")
add_dollar(g)
```

**Ce premier graph et les moyennes de chaque catégorie nous montrent qu'en effet les personnes les mieux payées recherchent moins ou passivement**

Pour travailler avec un heat-map on va d'abord changer le genre et lf_jobs par un booleen

```python
df.loc[(df["lf_jobs"] == "No"), "lf_jobs"] = 0
df.loc[(df["lf_jobs"] == "Yes, but only passively (just curious)"), "lf_jobs"] = 0
df.loc[(df["lf_jobs"] == "Yes, actively looking for something else"), "lf_jobs"] = 1
df["lf_jobs"] = df["lf_jobs"].astype("float")

df.loc[(df["gender"] == "Female"), "gender"] = 0
df.loc[(df["gender"] == "Male"), "gender"] = 1
df["gender"] = df["gender"].astype("float")


plt.figure(figsize=(6, 6))
g = sns.heatmap(df.corr())
```

```python
df.corr() 
```

**La correllation n'est pas significative l'hypothèse doit donc être rejetée**

D'une manière globale on ne trouve aucune corrélation avec le genre ce qui signifie que les tendances observées ne sont pas significatives. Pour confirmer ou infirmer ces tendances il faudrait un échantillion plus important notemment dans les hauts salaires.
