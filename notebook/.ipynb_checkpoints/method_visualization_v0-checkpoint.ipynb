## Fonctions de formatage (dans un fichier à part)

```python
# Function for creating barplot
def make_barplot(title, series, size=(10,6)):
    plt.figure(figsize=size)
    plt.title(title)
    return sns.barplot(x=series.index, y=series.values)

# Function for formatting salary on "xK $"
def to_k_usd(x):
    return str(int((x / 1000))) + "K $"
```

### Variable quantitatives

#### Principaux indicateurs

Méthode rapide

```python
round(df1["SalaryUSD"].describe())
```

La méthode propre pour les rendus

```python
"""
Présenter vos résultats mêmes simples par une phrase est une bonne pratique
Utiliser "print" plutot qu'un simple commentaire permet de mettre à jour facilement
vos analyses
""" 

salary_mean = df1["SalaryUSD"].mean()
salary_med = df1["SalaryUSD"].median()
salary_var = df1["SalaryUSD"].var(ddof=0)
salary_std = df1 ["SalaryUSD"].std(ddof=0)
salary_cv = salary_std / salary_mean

print("Le salaire annuel moyen aux USA est de {} et le salaire médian est de {}".format(to_k_usd(salary_mean), to_k_usd(salary_med)))
print("Son écart type des de {}".format(to_k_usd(salary_std)))
print("Le coefficient de variation de Salary USD est {:.2f}".format(salary_cv))                                                                                                              
                                                                                                                  
```

Les indicateurs de forme

Le Skewness ou coefficient d'asymétire: le cube des écarts à la moyenne diviser par le cube de l'écar type:
- Si Skewness=0 alors la distribution est symétrique.
- Si Skewness>0 alors la distribution est étalée à droite (une majorité des valeurs sont supérieures à la moyenne)
- Si Skewness<0 alors la distribution est étalée à gauche. (une majorité des valeurs sont inférieures à la moyenne)

```python
salary_skn= df1["SalaryUSD"].skew()

print("Le Skewness de  Salary USD est {:.2f}".format(salary_skn))       
```

Le kustosis ou coefficient d'aplatissement 
- Si Kurtosis=0 , alors la distribution a le même aplatissement que la distribution normale.
- Si Kurtosis>0 , alors elle est moins aplatie que la distribution normale : les observations sont plus concentrées.
- Si Kurtosis<0 , alors les observations sont moins concentrées : la distribution est plus aplatie.

```python
salary_kt= df1["SalaryUSD"].kurtosis()

print("Le Kurtosis de  Salary USD est {:.2f}".format(salary_kt))       
```

## Les liens entre deux variables

### Coefficient de corrélation linéaire

Le coefficient de corrélation linéaire (ou de Pearson) correspond à la covariance divisée par le produit des écarts types.
- Si le coefficient de corrélation = 0: il n'y a pas de corrélation linéaire: les deux phénomènes sont linéairement indépendant
- Si le coefficient de corrélation > 0: il y a une corrélation linéaire positive entre les deux valeurs
- Si le coefficient de corrélation < 0: il y a une corrélation linéaire négative (quand l'une croit, l'autre décroit)

```python
import scipy.stats as st
import numpy as np

print(st.pearsonr(df1["YearsWithThisTypeOfJob"],df1["SalaryUSD"])[0])
```

### Lien entre deux variables quantitatives continues

La première solution est d'afficher un diagramme de dispersion

```python
plt.plot(df1["YearsWithThisTypeOfJob"],df1["SalaryUSD"],'o',alpha=0.5)
plt.xlabel("Nb années dans ce type d'emploi")
plt.ylabel("Salaire")
plt.show()
```

Le résultat n'est pas très lisible. Je peux choisir de discrétiser ma variable continue

### Lien entre deux variables quantitatives continues et une variables quantitatives discrètes

#### Discrétisation d'une variable conitnue

L'objectif va être de créer des classes, je souhaite que ces classes à peu près équivalentes en terme d'effectif, je dois donc représenter la distribution de ma variable continu pour pouvoir la discrétiser.

```python
df1["YearsWithThisTypeOfJob"].hist(density=True,bins=20)
plt.show()
```

```python
df1['YearsWithThisTypeOfJob_b']=pd.cut(x=df1["YearsWithThisTypeOfJob"], bins=[0,1,2,3,5,10,42])
# Affichage de la distribution d'une variable quantitative discrète
s = df1['YearsWithThisTypeOfJob_b'].value_counts(normalize=True).sort_index(ascending=True)
s.plot(kind='bar')
plt.show()
```

#### Croisement d'une variable continue et discrète

```python
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="YearsWithThisTypeOfJob_b", y="SalaryUSD",palette=["g"],
            data=df1)
sns.despine(offset=10, trim=True)
```

### Lien entre une variable Numérique et une variable catégorielle

```python
modalites = df1["Gender"].unique()
groupes = []
for m in modalites:
    groupes.append(df1[df1["Gender"]==m]["SalaryUSD"])

# Propriétés graphiques (pas très importantes)    
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}
    
plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
            vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
plt.show()
```
